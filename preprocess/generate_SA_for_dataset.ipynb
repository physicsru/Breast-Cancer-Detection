```python
import os, sys, time
import cv2
import numpy as np
from skimage import exposure
sys.path.append("../pyusct/")
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
LOCAL_PATH = "/mnt/nas/"
SA_DIR = os.path.join(LOCAL_PATH, "nb_usctsim/sim_015/SA/")
GT_DIR = os.path.join(LOCAL_PATH, "nb_usctsim/sim_015/GT/")
DATASET_DIR = os.path.join(LOCAL_PATH, "PYUSCT_train/dataset036/")
LABEL_DIR = os.path.join(DATASET_DIR, "output/")
OUT_DIR = os.path.join(DATASET_DIR, "SA/")
```

```python
import pandas as pd
```

```python
df = pd.read_csv(os.path.join(DATASET_DIR, "list.csv"))
```

```python
pre_trial = None
SA = None
i = 0
for index, row in df.iterrows():
    trial = row["source"].split("/")[-1]
    ix, iy = row["ix"], row["iy"]
    if trial != pre_trial:
        SA_path = os.path.join(SA_DIR, trial+'.bmp')
        SA = cv2.imread(SA_path, 0)
        GT_path = os.path.join(GT_DIR, trial+'.bmp')
        GT = cv2.imread(GT_path, 0)
        pre_trial = trial
        i = 0
        
    vSA = SA[iy, ix]
    vGT = GT[iy, ix]
    file_name = trial + "_1batch_" + str(i).zfill(4) + ".npy"
    print(file_name)
    np.save(os.path.join(OUT_DIR, file_name), vSA)
    label = np.load(os.path.join(LABEL_DIR, file_name))
    i += 1
    print(trial, i, iy, ix, label, vGT, vSA)
```
